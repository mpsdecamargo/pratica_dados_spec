# Exercicio 1
Crie um Select que gere os 10 maiores Pesos do IBOV.

```python
#%pip install sqlalchemy pandas
```

```python
from sqlalchemy import create_engine, String, Float
import pandas as pd
import os
import sqlite3
```

```python
df1 = pd.read_csv("tabela_ex1_sql.txt", sep="\t")
df1.head(50)
```

```python
df1 = df1.set_index("ID")
df1.head(50)
```

```python
df1.info()
df1["QTD_TEORICA"] = df1["QTD_TEORICA"].str.replace(".", "").str.replace(",", ".").astype(float)
df1["PESO"] = df1["PESO"].str.replace(",", ".").astype(float)
```

```python
df1.info()
```

```python
DATABASE_PATH = os.path.join('db', 'sql_database.db')
SQLALCHEMY_DATABASE = f'sqlite:///{DATABASE_PATH}'
engine = create_engine(SQLALCHEMY_DATABASE)
print(SQLALCHEMY_DATABASE)

sql_script_path= "script_criar_db.sql"
```

```python
def criar_banco_de_dados():
    if not os.path.exists(DATABASE_PATH):
        try:
            conn = sqlite3.connect(DATABASE_PATH)
            cursor = conn.cursor()
            with open(sql_script_path, 'r') as f:
                sql_script = f.read()
            cursor.executescript(sql_script)
            conn.commit()
            print("Banco de dados criado.")
        except Exception as e:
            print(f"Erro: {e}")
        finally:
            conn.close()
    else:
        print("Banco de dados j√° existe.")
```

```python
criar_banco_de_dados()
```

```python
formato_colunas = {'TICKER': String(5), 'INDICE': String(50), 'FONTE': String(50), 'QTD_TEORICA': Float(), 'PESO': Float()}
db = df1.to_sql('INDEX_WEIGHTS', con=engine, if_exists='append', index=False, dtype=formato_colunas)
```

```python
conn = sqlite3.connect('db\\sql_database.db')

query = "SELECT * FROM INDEX_WEIGHTS"

df_sql = pd.read_sql(query, conn)
conn.close()
df_sql.head(50)
```

```python
query = "SELECT ID, TICKER, PESO FROM INDEX_WEIGHTS \
        WHERE INDICE LIKE 'IBOV' \
        ORDER BY PESO DESC \
        LIMIT 10"

df_sql_ex1 = pd.read_sql(query, conn)
df_sql_ex1.head(50)
conn.close()
```
